```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA

data = load_iris()
x = data.data
y = data.target

print(x.shape)
```

```python
scaler = StandardScaler()
x_standardized = scaler.fit_transform(x)
```

```python

lda = LinearDiscriminantAnalysis(n_components=2)
x_lda = lda.fit_transform(x_standardized, y)

plt.figure(figsize=(8, 6))
for class_label, marker, color in zip(range(3), ['o', 's', '^'], ['red', 'green', 'blue']):
    plt.scatter(x_lda[y == class_label, 0], x_lda[y == class_label, 1],
                label=data.target_names[class_label],
                marker=marker,
                color=color)
plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
plt.legend()
plt.title("LDA: 2D Projection of Iris Dataset")
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_standardized)

plt.figure(figsize=(8, 6))
for class_label, marker, color in zip(range(3), ['o', 's', '^'], ['red', 'green', 'blue']):
    plt.scatter(X_pca[y == class_label, 0], X_pca[y == class_label, 1],
                label=data.target_names[class_label],
                marker=marker,
                color=color)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.title("PCA: 2D Projection of Iris Dataset")
plt.show()
```
